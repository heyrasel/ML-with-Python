```python
# price ta numpy te array hisebe rekhe use korle
price_np = np.array([100,250,380,60])
discounted_price = price_np * 0.8 # emn mathematical operations kora jai
print(discounted_price) 
```

```python
empty_tables = np.zeros((2,3)) # mane 2 ta row and 3ta column er 2D array create hobe jar sob value 0
print(empty_tables)
```

```python
full_tables = np.ones((2,3)) # uporer tar ulta kaj holo, matrix er sob datar value 1 hoye gelo.
print(full_tables)
```

```python
bbq_boxes = np.full((2,3), 'BBQ') # BBQ dara sob value fill korlam
print(bbq_boxes)
```

```python
bbq_boxes[0,1] = 'VEGETABLE' # 0 row and 1 column er value change koralm.
print(bbq_boxes)
```

```python
# Let random kono value generate korbo 2,3 matrix form e.
random_discount = np.random.rand(2,3) # rand--> 0--1 er moddhe value generate kore randomly
print(random_discount)
```

```python
# Randomly integer number generate korte
random_int = np.random.randint(1,8, (2,3)) # mane 1--8 er moddhe number generate hobe. and 2,3 holo 2D array size.
print(random_int)
```

```python
# Identity matrix generation
identity_matrix = np.identity(3) # 3*3 identity matrix generate korbe.
print(identity_matrix)
```

```python
# Interval ba kichu number por por matrix er value chaile 
delivery_times = np.arange(0,121,15) # mane 0 theke 15 interval por por ekta value pabo upto 120 porjonto
print(delivery_times)
```

```python
# Specific vag e vag kore interval 
specific_interval = np.linspace(0,60,8) # 0 theke start and 60 porjonto and 8 vage vag hobe.
print(specific_interval)
```

```python
# Creating 2D array/ list inside another list
orders = np.array(
    [
        ['BBQ','CHEESE', 'CHICKEN'],
        ['VEG', 'SPICY', 'BEEF']
    ]
)
print(orders)
```

```python
print(orders[1,0])
```

```python
print(orders[1: ]) # mane 1st row er sob column dekhabe. egula ke slicing bola hoy.
```

```python
print(orders[ :, 1]) # only 1number column dekhte.
```

```python
print(orders[1, 0:2]) # 1 number row er 0 theke 1 column porjonto dekhabe. comma deya mane oi column er sob.
```

```python
print(orders[: , 1:]) # sobgula column and 1st col theke sobgula col
```
