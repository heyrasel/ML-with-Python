```python
# A Pandas DataFrame is a two-dimensional, size-mutable, tabular data structure with labeled axes (rows and columns).
# Mainly csv file theke data read kore dataframe e rekhe kaj kori but ajke amra nijera ekta dictionary create kore then kaj korbo.

import pandas as pd
data = {
    'Name' : ['Pandy', 'Bamboo', 'Fuzzy'],
    'Age' : [12, 4, 5],
    'Bamboo_Eat_Daily': [45, 53, 12]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Let age 6 bochor ba tar kom jader tader data gula dekhbo, so ekhn filtering korte hobe.
df[df['Age']<=6] # mane df er data ta filtering kortechi and [er moddher ta holo condition] . evabe column er upor filtering korte hoy.
```

```python
df[df['Name']=='Bamboo'] # only jader nam bamboo tader ta dekhabe.
```

```python
# Left,right join on dataframe

left = pd.DataFrame(
    {
        'ID': [1, 2, 3],
        'Name': ['Bamboo', 'Fuzzy', 'Pandy']
    }
)

right = pd.DataFrame(
    {
        'ID': [1, 2, 4],
        'Hobby': ['Sleeping', 'Eating', 'Sitting']
    }
)

merged = pd.merge(left, right, on= 'ID', how='right') # kiser upor join korbo eta on e dite hobe and kivabe join korchi left na right seta how e dite hobe
merged
```
