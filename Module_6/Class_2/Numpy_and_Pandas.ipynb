```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        'Name': ['Rasel', 'Ali', np.nan, 'Ahmed', 'Mahmud'], # np.nan means emtpty value
        'Age': [25,26,np.nan,27,22],
        'Salary': [60000,52000,48000,np.nan,48000]
    }
)
df
```

```python
# if we want to remove the NaN value 
# But within 5 data 2 data drop is risky.
data = df.dropna()
data
```

```python
# So we will fill the NaN values.
data = df.fillna(
    {
        'Name': 'Unknown', # ba jekono nam use korte pari
        'Age': df['Age'].mean(),
        'Salary': 0 # ba mean salary dite pari
    }
)
data
```

```python
# Age column ke integer e convert korte.
data['Age'] = data['Age'].astype(int)
data
```

```python
# We use 2D matrix for image training. like CNN.

np_array = np.array([[1,2,3],[4,5,6]]) # this is 2D array(list within list)
print(np_array)
```

```python
# Some other model uses 1D matrix for training. So we need reshape the matrix
reshape_array = np_array.reshape(3,2)
print(reshape_array)
```

```python
# But reshape korar somoy row*col = value jeta oi hisebe reshpaing korte hobe.
# Like, original (3,2) = 6 so reshaping could possible to (6,1), (2,3), (1,6) and so on.
reshape_array = np_array.reshape(1,6)
print(reshape_array)
```

```python
# Sometimes we have to join two matrix.
# If we join row wise then it call v_stack

a = np.array([[1,2,3],[4,5,4]])
b = np.array([[6,7,8],[9,10,11]])
```

```python
a
```

```python
b
```

```python


# But join korar somoy kheyal rakhte hobe jeno column size same thake 2tar e.
v_stacked = np.vstack([a,b])
print(v_stacked)
```

```python
# Horizontal stacking e column thik thake but row add hoye jai
h_stacked = np.hstack([a,b])
print(h_stacked)

```

```python
# d stack or depthwise stack e 2ta matrix er e row and column same hoyte hoy.
# 1st matrix er (1,1) er value and 2nd matrix er (1,1) er value mile ekta row hoy evabe.
# ekta 3D matrix create hoy.

d_stacked = np.dstack([a,b])
print(d_stacked)
```
