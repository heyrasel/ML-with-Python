```python
# Model training korar jonno kono column ke amra 0 ba 1 e convert korte pari map use kore.
# here amra konta kon value hobe ischamoto assign kore dite pari map er maddhome. 
df['Sex_Number'] = df['Sex'].map({'male':0, 'female':1})
df.head()

# ekhon model train korar somoy sex e 0 ba 1 diye model train korate pari.
```

```python
# Missing value kon kon column e ache eta check korbo ekhon.
missing_per_column = df.isnull().sum()
print(missing_per_column)
```

```python
# Now Embarked column e 2ta missing value ache. etogula row theke 2ta row baad dile kono problem hobe na so Embarked er jei 2ta row te NaN ache oi 2ta row delete kore dibo.

df = df.dropna(subset=['Embarked']) 
df.shape
```

```python
# lets check missing value deleted or not in Embarked column.
missing_per_column = df.isnull().sum()
print(missing_per_column)
```

```python
# Now Cabin column e missing value ache 687 ta. so etogula row delete kora jabe na.
# So Cabin column tai pura delete kore dibo. because ei column ta not that much important.
df = df.drop('Cabin', axis=1)
```

```python
missing_per_column = df.isnull().sum()
print(missing_per_column)
```

```python


df['Age'].mean()
```

```python
# Now age column e 177 ta value missing.
# So we cant drop 177 value and also Age column is important.
# So we will fill the NaN value of age column using mean value of the age.
# We will see future in details when we will fill by mean,meadian,mode.

df['Age'] = df['Age'].fillna(df['Age'].mean()) # mane Age er missing value gula ke age er mean diye replace kore dilam.

missing_per_column = df.isnull().sum()
print(missing_per_column)
```

```python
# Ekhon ei modified data ta save korte
df.to_csv('modified_train.csv')
```
